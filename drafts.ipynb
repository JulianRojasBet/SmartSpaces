```python
from urllib.request import urlopen
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_img_from_url(url):
    req = urlopen(url)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv2.imdecode(arr, -1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    print(img.shape)
    return img

url = 'https://a0.muscache.com/im/pictures/miso/Hosting-33996825/original/ae88e614-cd39-4503-b16c-e2a7a98556f5.jpeg?im_w=960'
img = read_img_from_url(url)
plt.figure(figsize=(4,4))
plt.imshow(img)
```

```python
# 0 -> black | 255 -> white
def draw_rectangle_on_image(img, x1, y1, x2, y2):
    img = cv2.rectangle(img, (x1, y1), (x2, y2), (255, 255, 255), -1)
    return img

# modified = draw_rectangle(img, 300,300, 200, 200)
modified = draw_rectangle_on_image(img, 300,300, 600, 100)
plt.figure(figsize=(4,4))
plt.imshow(modified)
```

```python
import numpy as np
import cv2

def get_rectangle_mask(x1, y1, x2, y2, shape):
    """
    Returns a rectangular mask of given dimensions and shape

    Arguments:
    x1, y1, x2, y2 - Coordinates of the rectangle
    shape - Shape of the mask image. Tuple representing height and width of the image

    Output:
    rect_mask - A rectangular mask of the given dimensions
    """
    rect_mask = np.zeros(shape, dtype=np.uint8)
    rect_mask = cv2.rectangle(rect_mask, (x1, y1), (x2, y2), (255, 255, 255), -1)
    rect_mask = cv2.cvtColor(rect_mask, cv2.COLOR_BGR2RGB)
    return rect_mask

rectangle = get_rectangle_mask(100, 100, 300, 300, (400, 400))
plt.figure(figsize=(4,4))
plt.imshow(rectangle)
```
